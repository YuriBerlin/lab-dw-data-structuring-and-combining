```python
import pandas as pd
import re
```

```python
df_1 = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv")
```

```python
df_2 = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv")
```

```python
df_3 = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv")
```

```python
df_1.columns = df_1.columns.str.lower().str.replace(" ", "_")
df_1.head()
```

```python
df_2.columns = df_2.columns.str.lower().str.replace(" ", "_")
df_2.head()
```

```python
df_3.columns = df_3.columns.str.replace("State", "ST")
df_3.columns = df_3.columns.str.lower().str.replace(" ", "_")
df_3.head()
```

```python
df = pd.concat([df_1, df_2, df_3])
df.reset_index(drop=True, inplace=True) 
df.head()
```

```python
df.head()
```

```python
len(df_1)
```

```python
len(df_2)
```

```python
len(df_3)
```

```python
len(df)
```

```python
df.gender.unique()
```

```python
df.gender = df.gender.apply(lambda i:str(i))
for i, value in enumerate(df['gender']):
    x = re.search(r'^F.*|^f.*', value)
    y = re.search(r'^M.*|^m.*', value)
    if x:
        df.at[i, 'gender'] = 'F'
    elif y:
        df.at[i, 'gender'] = 'M'  
    else:
        df.at[i, 'gender'] = 'unknown'
```

```python
df.gender.unique()
```

```python
df.st = df.st.apply(lambda i:str(i))
df.st.unique()
```

```python
state_mapping = {'Cali':'California', 'AZ':'Arizona','WA': 'Washington', 'nan': 'unknown'}
df.st = df.st.replace(state_mapping)
df.st.unique()
```

```python
df.education = df.education.apply(lambda i:str(i))
df.education.unique()
```

```python
edu_mapping = {'Bachelors':'Bachelor', 'nan': 'unknown'}
df.education = df.education.replace(edu_mapping)
df.education.unique()
```

```python
df.customer_lifetime_value = df.customer_lifetime_value.apply(lambda i:str(i))
df.customer_lifetime_value = df.customer_lifetime_value.apply(lambda i: i.replace('%',''))
```

```python
df.customer_lifetime_value = pd.to_numeric(df.customer_lifetime_value, errors='coerce')
```

```python
df.head()
```

```python
df.vehicle_class = df.vehicle_class.apply(lambda i:str(i))
car_mapping = {"Sports Car": "Luxury", "Luxury SUV" :"Luxury", "Luxury Car" : "Luxury", "nan" : "unknown"}
df.vehicle_class = df.vehicle_class.replace(car_mapping)
df.vehicle_class.unique()
```

## Bonus

You work at the customer service department and you want to know which months had the highest number of complaints by policy type category. Create a summary table showing the number of complaints by policy type and month.
Show it in a long format table.

*In data analysis, a long format table is a way of structuring data in which each observation or measurement is stored in a separate row of the table. The key characteristic of a long format table is that each column represents a single variable, and each row represents a single observation of that variable.*

*More information about long and wide format tables here: https://www.statology.org/long-vs-wide-data/*

```python
sub_df = analysis[['policy','total_claim_amount','month']]
sub_df.head(10)
```

```python
sub_df.columns
```

```python
melted_data = pd.melt(sub_df, id_vars=['policy', 'month'])
melted_data.head()
```

```python
melted_data[['policy', 'month', 'value']].sort_values(by=['policy','month']).head(10)
```

```python
melted_data_agg = melted_data.groupby(['policy','month']).agg({'value': 'sum'})
melted_data_agg
```
